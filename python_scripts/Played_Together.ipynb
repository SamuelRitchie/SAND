```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from collections import Counter
```

```python
df = pd.read_csv("data/all_data.txt", sep= " ", header=None, names=['Artiste1', 'Lien', 'Artiste2','.'])
metadata = pd.read_csv('data/metadata_dataframe.csv')
```

```python
set(df.Lien)
```

```python
def keep_row(lien):
    if 'collaborated_with' in lien or 'inBand' in lien or  'bandmember' in lien or "playedTogether" in lien or \
    'bandLeaderOf' in lien or 'touredWith' in lien or 'mentorOf' in lien:
        return True
    else:
        return False
df["Keep"] = df.Lien.apply(keep_row)
df = df[df.Keep == True]
```

```python
df = df[["Artiste1", 'Artiste2', 'Lien']]
corresp = metadata[["artist_url", 'Name']].set_index('artist_url').to_dict()
def url_to_artist(url):
    try:
        return corresp['Name'][url]
    except:
        return None

df.Artiste1 = df.Artiste1.apply(url_to_artist)
df.Artiste2 = df.Artiste2.apply(url_to_artist)
df = df[['Artiste1', 'Artiste2','Lien']]
df.dropna(axis=0, how='any')
df["Id"] = df.Artiste1+df.Artiste2
df = df.drop_duplicates(subset='Id')
df = df[['Artiste1', 'Artiste2']]
```
